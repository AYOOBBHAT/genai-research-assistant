```python
from transformers import pipeline
```

```python
generator = pipeline(
    task="text-generation",
    model="gpt2",
    device_map="auto"
)
```

```python
prompt ="Artificial Intelligence will impact society by"
```

```python
outputs=generator(
    prompt,
    max_new_tokens=200,
    num_return_sequences=1,
    do_sample=True,
    temperature=0.8,
    top_p=0.9,

)
for i,out  in enumerate(outputs):
  print(f"\n--- Output {i+1} ---\n")
  print(out["generated_text"])




```

```python
prompts = [
    "Explain AI simply.",
    "Explain AI as a researcher.",
    "Explain AI with strict factual tone."
]

for p in prompts:
    print("\nPROMPT:", p)
    print(generator(p, max_new_tokens=40)[0]["generated_text"])
```
